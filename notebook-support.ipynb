# 📓 Notebook Support

MyST supports native rendering of Jupyter Notebooks, allowing executable code, rich outputs, and narrative documentation to coexist in a single document.

This capability is especially useful for data engineering, data science, and technical documentation, where explanations and executable examples should remain synchronized.

```python
from pathlib import Path 
docs = Path("/workspace/docs")
```

```python
print("📚 Documentation Pages")

for file in docs.glob("*.md"):
    print(f"  - {file.name}")
```
